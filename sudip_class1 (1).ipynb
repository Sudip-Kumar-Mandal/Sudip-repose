```python
li = [1,2,3,4,5,6]
print(li)
print(type(li))

#array is a collection of homogenous values,
#list is a collection of heterogenous values(int,char,floar...)

print(li[0])

print(li[-1]) #first element from the end
print(li[-2])
```

```python
print(li)

#slicing operation: 

print(li[1:5])  #start from 1. go till 4(n-1)
print(li[2:])   # start from index 2 , go till end
print(li[:4])  # start from begining, go till 3(n-1)
print(li[:])
```

```python
print(li)
print(li[ : :1])  #li(start : end : step)
print(li[ : : 2])  # step by 2
print(li[::3])
print(li[ : : -1])  #printing in reverse order
```

```python
#LIST METHOD:

li = [1,2,3,4,5,6]

li.insert(0, 15) #insert(index, element you wanna insert)
print(li)

li.insert(20, 18)
print(li)

li.insert(-20, 'hello')
print(li)

li.pop()   #delet last element
print(li)

li.pop(0)
print(li)
```
